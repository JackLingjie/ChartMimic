```python
import json  
  
# 读取原始数据  
with open("/mnt/lingjiejiang/multimodal_code/data/chart_data/ChartBench/image_info_68k.json", 'r') as f:  
    raw_data = json.load(f)  
  
# 计算每份数据的大小  
total_length = len(raw_data)  
split_size = total_length // 4  
  
# 切分数据  
split_data_1 = raw_data[:split_size]  
split_data_2 = raw_data[split_size:2*split_size]  
split_data_3 = raw_data[2*split_size:3*split_size]  
split_data_4 = raw_data[3*split_size:]  
  
# 保存切分后的数据  
with open("/mnt/lingjiejiang/multimodal_code/data/chart_data/ChartBench/image_info_68k_split_1.json", 'w') as f1:  
    json.dump(split_data_1, f1)  
with open("/mnt/lingjiejiang/multimodal_code/data/chart_data/ChartBench/image_info_68k_split_2.json", 'w') as f2:  
    json.dump(split_data_2, f2)  
with open("/mnt/lingjiejiang/multimodal_code/data/chart_data/ChartBench/image_info_68k_split_3.json", 'w') as f3:  
    json.dump(split_data_3, f3)  
with open("/mnt/lingjiejiang/multimodal_code/data/chart_data/ChartBench/image_info_68k_split_4.json", 'w') as f4:  
    json.dump(split_data_4, f4)  
  
print("数据已成功切分并保存。")  
```

```python
import matplotlib.pyplot as plt

# Data for the inner ring
inner_labels = ['Housing', 'Transportation', 'Groceries', 'Entertainment', 'Personal Care']
inner_sizes = [30, 25, 20, 15, 10]
inner_colors = ['#cccccc', '#999999', '#666666', '#333333', '#000033']

# Data for the outer ring
outer_labels = [
    'Rent', 'Utilities', 'Maintenance', 'Gasoline', 'Public Transport',
    'Car Insurance', 'Maintenance', 'Fresh Produce', 'Canned Goods', 'Snacks',
    'Beverages', 'Dining Out', 'Streaming Services', 'Books/Magazines', 'Activities',
    'Toiletries', 'Hair Care', 'Skincare', 'Makeup'
]
outer_sizes = [
    16.7, 8.3, 4.4, 5.6, 3.9, 
    2.8, 1.7, 6.7, 4.4, 2.8,
    5.6, 8.3, 3.9, 2.8, 2.8,
    4.4, 5.6, 3.9, 5.6
]
outer_colors = [
    '#00ffff', '#00ced1', '#00bfff', '#1e90ff', '#4169e1',
    '#4682b4', '#5f9ea0', '#6495ed', '#7b68ee', '#6a5acd',
    '#8a2be2', '#9932cc', '#8b008b', '#9400d3', '#800080',
    '#8b4513', '#a0522d', '#cd5c5c', '#ff69b4'
]

fig, ax = plt.subplots(figsize=(10, 9))

# Create the inner circle
ax.pie(inner_sizes, radius=1.1, labels=inner_labels, colors=inner_colors, wedgeprops=dict(width=0.3, edgecolor='w'))

# Create the outer circle
ax.pie(outer_sizes, radius=1.4, labels=outer_labels, colors=outer_colors, labeldistance=1.08, wedgeprops=dict(width=0.3, edgecolor='w'))

# Title and display
plt.title('Distribution of Monthly Expenses by Category', fontsize=16)
plt.tight_layout()
plt.show()
```
